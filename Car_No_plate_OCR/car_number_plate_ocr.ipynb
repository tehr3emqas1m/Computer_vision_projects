```python
import easyocr
from google.colab import files
from IPython.display import Image, display
import cv2
import numpy
from google.colab.patches import cv2_imshow
```

```python
def detect_license_plate(image):
  reader = easyocr.Reader(["en"])
  result = reader.readtext(image)
  license_plate_text = ""
  for detection in result:
    text = detection[1]
    if any(char.isdigit() for char in text) and any(char.isalpha() for char in text):
      license_plate_text = text
  return license_plate_text

def main():
  uploaded = files.upload()
  image_filename = list(uploaded.keys())[0]
  display(Image(image_filename))
  license_plate_text = detect_license_plate(image_filename)
  if license_plate_text:
    print("The detected car license plate is:", license_plate_text)
  else:
    print("No license plate detected")

if __name__ == "__main__":
  main()
```
